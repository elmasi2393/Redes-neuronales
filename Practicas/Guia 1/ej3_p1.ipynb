```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots

plt.style.use(['science', 'grid', 'notebook'])
```

```python
# retorna el valor de la ecuaci√≥n de Goldman
K = 8.6e-5   # constante de Boltzmann [eV/K]
def ja(Ain, Aout, V, T, na, Pa):
    return Pa * na * V /(K*T) * (Ain * np.exp(V*na/(K*T)) - Aout) / (np.exp(V*na/(K*T)) - 1)
```

```python
concentrations = {'ion': ['K', 'Na', 'Cl'], 'in': [430, 50, 65], 'out': [20, 440, 550], 'valence': [1, 1, -1]}
V = np.linspace(-400e-3, 400e-3, 1000)
Pa = 1

fig, ax = plt.subplots()

for i in range(3):
    ax.plot(V, ja(concentrations['in'][i], concentrations['out'][i], V, 310, concentrations['valence'][i], Pa), label=concentrations['ion'][i])

ax.set_xlabel('V [V]')
ax.set_ylabel('j [A/m¬≤]')
ax.legend()
plt.show()


```
